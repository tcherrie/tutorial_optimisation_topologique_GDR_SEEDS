# Utilisation d'un notebook Jupyter



# 1) Qu'est-ce qu'un notebook Jupyter ?

Jupyter est un outil gratuit et puissant qui permet de créer des **notebooks**, c'est-à-dire un document qui contient à la fois du contenu textuel et du code exécutable, lisible par un navigateur internet. Il s'agit d'un format très pratique pour l'enseignement collaboratif. Un notebook est organisé en **cellules**. On peut repérer l'étendue d'une cellule en cliquant dessus (une fois). Il en existe deux types principaux, qui peuvent contenir :
- soit du **code exécutable**
- soit du **texte**, écrit en *markdown*

# 2) Ecrire et exécuter du code Python3

Commençons par écrire un bout de code :

```python
print("Hello world !")
```

La *cellule* ci-dessus s'exécute lorsqu'on la sélectionne en cliquant dessus, puis qu'on appuie sur Maj + Entrée.
La sortie s'affiche en dessous de la cellule.

On déclare maintenant une variable. Exécutez la cellule ci-dessous :

```python
a = 1
```

Une fois le code d'une cellule exécuté, les variables restent en mémoire. En effet, si on exécute :

```python
print(a)
```

On peut donc définir des fonctions, importer des modules... Le tout est d'exécuter les cellules *dans l'ordre* ! Il est également possible de réexécuter les cellules autant de fois qu'on le souhaite.

# 3) Ecrire et exécuter du code Python3

Pour lui faire reprendre son aspect d'affichage, vous pouvez l'exécuter comme une cellule de code en appuyant sur Maj + Entrée.
